```python
### Imports 

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import os
```

```python
os.chdir('../Datasets')
os.environ["WANDB_API_KEY"] = "0"  # Silences warnings
```

Read the data

```python
train = pd.read_csv("train.csv")
train.head()
```

Count the different values in the language column and draw them as a pie of percentages

```python
labels, frequencies = np.unique(train.language.values, return_counts = True)
plt.figure(figsize = (10,10))
plt.pie(frequencies,labels = labels, autopct = '%1.1f%%')
plt.show()
```

Count the different values in the language column and draw them as a pie of percentages

```python
labels, frequencies = np.unique(train.label.values, return_counts = True)
plt.figure(figsize = (10,10))
plt.pie(frequencies,labels = labels, autopct = '%1.1f%%')
plt.show()
```
